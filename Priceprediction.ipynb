***Mobile price dataset*** is given and we need to train the model and make ready for new testing.

```python
#@title #####**Loading the dataset**
import pandas as pd
import matplotlib.pyplot as plt

df1=pd.read_csv('price.csv')
df1.head()
```

```python
#@title #####**Preprocessing with the dataset**

df=df1.drop(['Brand','Model','Camera (MP)'],axis=1)
df['Storage'] = df['Storage'].str.replace("GB", "")
df['RAM'] = df['RAM'].str.replace("GB", "")
df['Price ($)'] = df['Price ($)'].str.replace("$", "")
# df['RAM'] = df['RAM'].astype('float')

for col in df.columns:
 df[col] = pd.to_numeric(df[col], errors='coerce')
 df[col]=df[col].astype(float)


df['Price ($)']=df['Price ($)'].fillna(df['Price ($)'].mean())
df['Screen Size (inches)']=df['Screen Size (inches)'].fillna(df['Screen Size (inches)'].mean())


df.shape
# df.isnull().sum()
```

```python
#@title #####**Break dataset into training and testing data**

X=df.iloc[:,:-1]
Y=df.iloc[:,-1:]

# 407 ka (50%=204,60%=244,70%=285,80%=325)
n=325
xtrain=X[:n]
ytrain1=Y[:n]
xtest=X[n:]
ytest1=Y[n:]

ytrain=ytrain1.values.ravel();
ytest=ytest1.values.ravel();
print(len(xtrain),len(xtest))
df.head()


# len(df['Price ($)'].unique())
```

```python
#@title #####**Using LinearRegression model**

from sklearn.linear_model import LinearRegression
model =  LinearRegression()
model.fit(xtrain,ytrain)
print("Accuracy ",model.score(xtest, ytest))
```

```python
#@title #####**Comparing both prediction and test label dataset**
p=model.predict(xtest).astype('int')
print("prediction",p)
print("\nytest",ytest)
# model.predict([[128,6,6.2,3100]])
```
