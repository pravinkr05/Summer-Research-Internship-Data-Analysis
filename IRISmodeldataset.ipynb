**IRIS Dataset:** The iris data are a data frame of 150 measurements of iris petal and sepal lengths and widths, with 50 measurements for each species of “setosa,” “versicolor,” and “virginica.”

```python
#@title #####**Loading the dataset**
import pandas as pd
import seaborn as sns   #seaborn provides function that help to quick access to small dataset
sns.get_dataset_names() #show all dataset  available
iris=sns.load_dataset('iris') #loaded iris dataset
# iris.isnull().values.any()   # In entire dataset contains NAN value or not
iris.isnull().sum()  #column wise NAN value showing
iris.head()
```

```python
#@title #####**Counting the classes**
iris['species'] = iris.species.map( {'setosa':0.0 , 'versicolor':1.0, 'virginica':2.0} )
class_count=iris['species'].value_counts()
print(class_count)
```

```python
#@title #####**Sort and Print the dataset**
iris1=iris.sort_values('species',ascending=True)
iris1
```

```python
#@title #####**Break whole dataset into training data and testing data**

ds_0=iris1[:50]   #ds_0 === dataset with label=0
ds_1=iris1[50:100]
ds_2=iris1[100:150]
#ds_2.shape #no. of rows in dataset

ds_0_train=ds_0[:25]
ds_0_test=ds_0[25:]

ds_1_train=ds_1[:25]
ds_1_test=ds_1[25:]

ds_2_train=ds_2[:25]
ds_2_test=ds_2[25:]


frames1 = [ds_0_train,ds_1_train,ds_2_train]
train_ds = pd.concat(frames1)
# train_ds


frames2 = [ds_0_test,ds_1_test,ds_2_test]
test_ds = pd.concat(frames2)
# test_ds


train_ds_x=train_ds.iloc[:,:-1]
train_ds_y=train_ds.iloc[:,-1:]
# train_ds_y


test_ds_x=test_ds.iloc[:,:-1]
test_ds_y=test_ds.iloc[:,-1:]
# test_ds_x


train_x=train_ds_x;
train_y=train_ds_y.values.ravel();
test_x=test_ds_x;
test_y=test_ds_y.values.ravel();
# train_ds_y
print(len(train_x),len(test_x))
```

####**Using** **SVM**

```python
#using support vector machine

from sklearn.svm import SVC # "Support vector classifier"
model = SVC(kernel='linear')
model.fit(train_x, train_y)
predictions = model.predict(test_x)
print(predictions)
print(test_y)
```

```python
percentage = model.score(test_x, test_y)
print("percentage : ",percentage)
```

```python
# confusion matrix shows the errors in the model performance in the form of a matrix, known as an error matrix [test_y vs predictions].

from sklearn.metrics import confusion_matrix
res = confusion_matrix(test_y, predictions)
print("Confusion Matrix") # first row for class=0 and total 10 prediction is successful, 2nd row for class=1 in which 8 predictions are correct and 2 predictions are wrong , 3rd for class=2 similar to row1
print(res)
print(f"Test Set: {len(test_ds_x)}")
print(f"Accuracy = {percentage*100} %")
```
